```python
k = []
for i in range(len(comment)):
  L=[]
  for j in Treatment_names:
    for lol in Treatment[j]:
      # Use .iloc to access rows by numerical index
      if lol in comment['token'].iloc[i]:
        L.append(j)
      else:
        L.append('ไม่สามารถระบุได้')
  L =list(set(L))
  if len(L) >= 2:
    L.remove('ไม่สามารถระบุได้')
  k.append(L)
comment['Treatment'] = k
```
